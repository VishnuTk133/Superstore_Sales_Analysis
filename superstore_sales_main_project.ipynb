# 2. Data Pre-processing            

*   Handling missing values
*   Removing duplicates
*   Correcting data types
*   Creating derived columns
*   Data Filtering or aggregating data



```python
# Identified null values using .isnull().sum()
df.isnull().sum()
```

```python
# Checked duplicates using .duplicated()
df.duplicated()
```

```python
# detect count of duplicate values from data frame
df.duplicated().sum()
```

```python
# Check data types of all columns to change data types.
df.dtypes
```

```python
# rename column names
df.rename(columns={
    'Country/Region': 'Country',
    'State/Province': 'State',
    }, inplace=True)
```

```python
# create month and year column
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
```

```python
# creating derived columns
df['Total_Sales'] = df['Sales'] * df['Quantity']
df['Total_Profit'] = df['Profit'] * df['Quantity']
df['Total_Discount'] = df['Discount'] * df['Quantity']
df['Unit_Price'] = df['Sales'] / df['Quantity']
df['Unit_Cost'] = df['Profit'] / df['Quantity']
df['Profit_Margin'] = df['Profit'] / df['Sales']
df['Discount_Amount'] = df['Discount'] * df['Quantity']
df['Discount_Rate'] = df['Discount'] / df['Quantity']
df['Region_Segment'] = df['Region'] + '-' + df['Segment']
```

```python
# fetch column names
df.columns
```

```python
# conditional filtering
df[df['Sales'] > 1000] # To extract sales records with values greater than 1000 (Find high-value sales records).
```

```python
#conditional filtering
#Find profitable Technology orders
df[(df['Category'] == 'Technology') & (df['Profit'] > 100)]
```

```python
# Analyze bulk purchases
df[df['Quantity'] > 5]
```

```python
# Year-wise sales analysis.
df[(df['Year'] == 2025)]
```

```python
# Find Low Sales and High Discount
df[(df['Sales'] < 100) & (df['Discount'] > 0.1)]
```

# **4.Visualizations**

```python
# import library for Visualizations
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# sales by category (Bar plot)
category_sales = df.groupby('Category')['Sales'].sum().reset_index()
plt.figure(figsize=(8,5))
sns.barplot(x='Category', y='Sales', data=category_sales)

plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

```python
# Monthly Sales Trend (Line chart)
# df['Order Date'] = pd.to_datetime(df['Order Date'])
monthly_sales = df.groupby(df['Month'])['Sales'].sum()

plt.figure(figsize=(12,6))
monthly_sales.plot(marker='o')

plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
# Sales Contribution by Segment (Pie Chart)
segment_sales = df.groupby('Segment')['Sales'].sum()

plt.figure(figsize=(6,6))
plt.pie(segment_sales, labels=segment_sales.index, autopct='%1.1f%%', startangle=90)

plt.title('Sales Distribution by Segment')
plt
```

```python
# Sales Distribution (Histogram)
plt.figure(figsize=(8,5))
sns.histplot(df['Sales'], bins=20, kde=True)

plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
# Profit by Category (Box Plot)
plt.figure(figsize=(8,5))
sns.boxplot(x='Category', y='Profit', data=df)

plt.title('Profit by Category')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.show()
```

```python
# Sales vs Profit (Scatter Plot)
plt.figure(figsize=(8,5))
sns.scatterplot(x='Sales', y='Profit', hue='Category' , data=df)

plt.title('Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

```python
# Correlation Analysis (Heatmap)
corr = df[['Sales','Profit','Discount',
           'Quantity','Profit_Margin',
           'Unit_Price']].corr()

plt.figure(figsize=(8,6))

sns.heatmap(corr,
            annot=True,
            cmap='coolwarm')

plt.title('Correlation Heatmap')
plt.show()
```

```python
# Orders by Ship Mode (Count Plot)
plt.figure(figsize=(8,5))

sns.countplot(x='Ship Mode', data=df)

plt.title('Orders by Ship Mode')
plt.xlabel('Ship Mode')
plt.ylabel('Count')
plt.xticks(rotation=15)
plt.show()
```

```python
# Interactive Sales by Region (Plotly Interactive Chart)
region_sales = df.groupby('Region')['Sales'].sum().reset_index()

fig = px.bar(region_sales,
             x='Region',
             y='Sales',
             color='Region',
             title='Regional Sales Analysis')

fig.show()
```

```python
# Multiple Charts in One Layout (Subplots)
fig, axes = plt.subplots(2,2, figsize=(12,10))

sns.histplot(df['Sales'], kde=True, ax=axes[0,0])
axes[0,0].set_title('Sales Distribution')

sns.boxplot(x=df['Profit'], ax=axes[0,1])
axes[0,1].set_title('Profit Boxplot')

sns.countplot(x='Category', data=df, ax=axes[1,0])
axes[1,0].set_title('Category Count')

sns.scatterplot(x='Sales', y='Profit', data=df, ax=axes[1,1])
axes[1,1].set_title('Sales vs Profit')

plt.tight_layout()
plt.show()
```
